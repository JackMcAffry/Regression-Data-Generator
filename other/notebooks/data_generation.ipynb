# Data generation
The whole datas will be saved in "results" variable

## Separated params
It is the easiest way to generate datasets. You don't need to instantiate classes, just put mathematical parameters and generation choices into the functions "lr_separated_params" or "log_separated_params"

### Linear Regression

```python
from model.simple_access.functions import Generator

results = Generator.lr_separated_params(seed=11,n_variables=5,error_variance=0,n_points=30,related_vars=3,n_scales=1,requested_mean=2,unuseful_vars=1,opt_level='float32') 

print("features:\n{}\n".format(results.features))
print("response:\n{}\n".format(results.response))
print("correlation matrix:\n{}\n".format(results.corr_matrix))
print("covariance matrix:\n{}\n".format(results.cov_matrix))
print("user: {}\n".format(results.gen_details.user))
print("os name: {}".format(results.gen_details.platform_name))
```

### Logistic Regression

```python
from model.simple_access.functions import Generator

results = Generator.log_separated_params(seed=11,n_variables=5,error_variance=0,n_points=30,related_vars=3,n_scales=1,requested_odds=2,unuseful_vars=1) 

print("features:\n{}\n".format(results.features))
print("response:\n{}\n".format(results.response))
print("correlation matrix:\n{}\n".format(results.corr_matrix))
print("covariance matrix:\n{}\n".format(results.cov_matrix))
print("user: {}\n".format(results.gen_details.user))
print("os name: {}".format(results.gen_details.platform_name))
```

## By params instance
At first, you have to instance a class containing mathematical parameters. Later, you have to put this instance, together with generation choices, into the functions "lr" or "log".

### Linear Regression

```python
from model.simple_access.functions import Generator
from model.implementation_lr.lr_params import LR_Params

params = LR_Params(seed=11,n_variables=5,error_variance=0,n_points=30,related_vars=3,n_scales=1,requested_mean=2,unuseful_vars=1)
results = Generator.lr(params, verbose=False) 

print("features:\n{}\n".format(results.features))
print("response:\n{}\n".format(results.response))
print("correlation matrix:\n{}\n".format(results.corr_matrix))
print("covariance matrix:\n{}\n".format(results.cov_matrix))
print("user: {}\n".format(results.gen_details.user))
print("os name: {}".format(results.gen_details.platform_name))
```

### Logistic Regression

```python
from model.simple_access.functions import Generator
from model.implementation_log.log_params import Log_Params

params = Log_Params(seed=11,n_variables=5,error_variance=0,n_points=30,related_vars=3,n_scales=1,requested_odds=2,unuseful_vars=1)
results = Generator.log(params) 

print("features:\n{}\n".format(results.features))
print("response:\n{}\n".format(results.response))
print("correlation matrix:\n{}\n".format(results.corr_matrix))
print("covariance matrix:\n{}\n".format(results.cov_matrix))
print("user: {}\n".format(results.gen_details.user))
print("os name: {}".format(results.gen_details.platform_name))
```
